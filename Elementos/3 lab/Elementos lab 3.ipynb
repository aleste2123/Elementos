# Parte 3

Coneccion de un generador y un motor(trif√°sica) Estrella.

Arranque directo -> conectar el motor y ver que pasa.



```python
T = np.array([0.0001,0.05, 0.1, 0.15, 0.2, 0.25, 0.3, 0.36, 0.4, 0.45, 0.5, 0.55])
velo = np.array([3455, 3435, 3405, 3376, 3334, 3300, 3251, 3134, 3115, 3041, 2951, 2838])
V = np.array([118.9, 118.7, 118.5, 118.8, 118.5, 118.7, 118.7, 118.5, 118.7, 118.6, 118.7, 118.7])
I = np.array([0.36, 0.4, 0.44, 0.5, 0.55, 0.61, 0.7, 0.8, 0.88, 0.96, 1.07, 1.2])
P_a = np.array([35.02, 39.9, 45.62, 52.1, 59.6, 65.3, 75.29, 85.6, 95.34, 103.6, 0.12*1000, 0.13*1000])
fp = np.array([0.823, 0.85, 0.876, 0.89, 0.9, 0.91, 0.915, 0.92, 0.918, 0.914, 0.908, 0.9])
P_out = 2*3.14/60*velo*T
```

```python
fig, axs = plt.subplots(2, 3, figsize=(14, 8))

axs[0,0].scatter(T, velo)
axs[0,0].set_xlabel("Torque [Nm]")
axs[0,0].set_ylabel("Velocidad [RPM]")

axs[0,1].scatter(T, V)
axs[0,1].set_xlabel("Torque [Nm]")
axs[0,1].set_ylabel("Voltaje [V]")

axs[0,2].scatter(T, I)
axs[0,2].set_xlabel("Torque [Nm]")
axs[0,2].set_ylabel("Corriente [A]")


axs[1,0].scatter(T, P_a)
axs[1,0].set_xlabel("Torque [Nm]")
axs[1,0].set_ylabel("Potencia Activa [W]")


axs[1,1].scatter(T, fp)
axs[1,1].set_xlabel("Torque [Nm]")
axs[1,1].set_ylabel("factor de potencia")


axs[1,2].scatter(T, P_out)
axs[1,2].set_xlabel("Torque [Nm]")
axs[1,2].set_ylabel("Potencia de salida [W]")

plt.subplots_adjust(hspace=0.3, wspace=0.3)
```

```python
e = P_out/ (P_a*3)*100
print(e)
plt.scatter(T,e)
plt.ylabel("Eficiencia porcentual [%]")
plt.xlabel("Torque [Nm]")
```

# Datos a excel


```python
df = pd.DataFrame({'T': T, 'Velocidad': velo, 'V': V, 'I': I, 'P_activa': P_a, 'Factor_potencia': fp, 'P_salida': P_out, "eficiencia":e})

# Guardar el DataFrame en un archivo Excel
df.to_excel('datos.xlsx', index=False)
```
